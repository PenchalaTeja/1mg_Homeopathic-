```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver import Keys
from selenium.webdriver.common.by import By
import time

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

driver = webdriver.Chrome()
driver.get("https://www.1mg.com/categories/homeopathy-57")
driver.maximize_window()
# time.sleep(3)

cancel_b = driver.find_element(By.XPATH, '//div[@class="UpdateCityModal__cancel-btn___2jWwS UpdateCityModal__btn___oMW5n"]')
if cancel_b.is_displayed():
    cancel_b.click()
    # time.sleep(2)


html=driver.find_element(By.TAG_NAME,'html')
for i in range(5):
    html.send_keys(Keys.PAGE_DOWN)
    # time.sleep(2)
# You might need to handle any popups or modals here

# Collect product links
product_links = []
page=1
while page <= 10:
    product_elements = driver.find_elements(By.XPATH, '//div[@class="style__product-box___liepi"]/a')
    for product_element in product_elements:
        link = product_element.get_attribute('href')
        product_links.append(link)


    page+=1
    next_b_b = driver.find_element(By.XPATH, '//li[@class="next"]/a')
    if next_b_b.is_displayed():
        next_b_b.click()
    time.sleep(2)


print(len(product_links))

Name=[]
Brand=[]
Ingredients=[]
Benifits=[]
Ratings=[]
No_of_Ratings=[]



# Visit each product page and scrape data
for product_link in product_links:
    driver.get(product_link)
    # time.sleep(3)


    # Scraping product name
    try:
        product_name_element = driver.find_element(By.XPATH, '//div[@class="row OtcPage__top-container___2JKJ-"]/div[3]/div/div/h1')
        product_name = product_name_element.text
        # print("Product Name:", product_name)
        Name.append(product_name)
    except:
        # print("NA")
        Name.append("NA")

    #  Scraping product price
    try:
        Brand_element = driver.find_element(By.XPATH, '//div[@class="row OtcPage__top-container___2JKJ-"]/div[3]/div/div[2]/a')
        brand = Brand_element.text
        # print("Product Brand:", brand)
        Brand.append(brand)
    except:
         print("NA")
        Brand.append("NA")

                                               
    try:
        K_B = driver.find_element(By.XPATH, '//div[@class="ProductHighlights__highlights-text___dc-WQ"]/ul/li')
        k_b=K_B.text
        # print("K_B:",k_b)
        Benifits.append(k_b)
    except:
        # print("NA")
        Benifits.append("NA")

    try:                                               
        K_I_ = driver.find_element(By.XPATH, '//*[@id="container"]/div/div/div[3]/div[1]/div[2]/div[2]/ul[1]')
        k_n_ = K_I_.text
        # print("K_I_:", k_n_)
        Ingredients.append(k_n_)
    except:
        # print("NA")
        Ingredients.append("NA")

    try:
        rating=driver.find_element(By.XPATH,'//div[@class="RatingDisplay__ratings-container___3oUuo"]/span')
        Rating=rating.text
        # print('Rating:',Rating)
        Ratings.append(Rating)
    except:
        # print("NA")
        Ratings.append("NA")


    try:
        no_of_rating=driver.find_element(By.XPATH,'//div[@class="ProductTitle__ratings___4MWF_"]/span')
        No_of_Rating=no_of_rating.text
        # print("No_:",No_of_Rating)
        No_of_Ratings.append(No_of_Rating)
    except:
        # print("NA")
        No_of_Ratings.append("NA")

    # description_div = driver.find_element(By.XPATH, '//div[@class="ProductDescription__description-content___A_qCZ"]')
    #
    # if description_div:
    #     strong_tags = description_div.find_elements(By.TAG_NAME, 'strong')
    #     Key_Ingredients = None
    #
    #     for strong_tag in strong_tags:
    #         if strong_tag.text == 'Key Ingredients':
    #             ul_element = strong_tag.find_element(By.XPATH, 'following-sibling::ul')
    #             Key_Ingredients = ul_element.text
    #             break
    #     print("Key Ingredients:", Key_Ingredients)

    # Add more code to scrape other data as needed
#
# Create a DataFrame
data = {
    'Name': Name,
    'Brand': Brand,
    'Key Ingredients': Ingredients,
    'Benifits':Benifits,
    'Rating':Ratings,
    'No_of_Rating':No_of_Ratings
}

df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('med_details_500.csv', index=True)

# Close the WebDriver
driver.quit()
```
