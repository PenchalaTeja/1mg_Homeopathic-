```python
# Importing libraries
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python

Name=[]
Sizes=[]
MRPs=[]
Prices=[]
URLs=[]

page=1

while page <=2:
    header={
        'Origin': 'https://www.1mg.com',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
    }
    url="https://www.1mg.com/categories/homeopathy/homeopathy-medicines-106?filter=true&pageNumber="+str(page)
    res = requests.get(url=url,headers=header)
    # print(res.status_code)

    if res.status_code==404:
        print("break")
        break
    else:
        # print('else')
        obj=soup(res.content,'html.parser')
        Box=obj.find_all('div',{'class':'style__product-box___liepi'})
        for i in Box:
            name=i.find('div', {'class':'style__pro-title___2QwJy'})
            if name:
                Name.append(name.text)
            else:
                Name.append("Na")

            size=i.find('div',{'class':'style__pack-size___2JQG7'})
            if size:
                Sizes.append(size.text)
            else:
                Sizes.append("Na")

            mrps=i.find('span',{'class':'style__discount-price___25Bya'})
            if mrps:
                MRPs.append(mrps.text)
            else:
                MRPs.append("NA")

            price=i.find('div',{'class':'style__price-tag___cOxYc'})
            if price:
                Prices.append(price.text)
            else:
                Prices.append("NA")
            
            mg_url=i.find('a',{'class':'style__product-link___UB_67'})
            if mg_url:
                b="https://www.1mg.com/"+mg_url.get('href')
                URLs.append(b)
            else:
                URLs.append("NA")
        print(page,"page=>",end=" ")
    page+=1
#     print(len(URLs),end=" ")
#     print(len(Prices),end=" ")
#     print(len(Sizes),end=" ")
#     print(len(MRPs),end=" ")
#     print(len(Name))


data={"name":Name,"size_of_bottle":Sizes,"MRPs":MRPs,"selling_price":Prices,"1mg_url":URLs}
# print(data)
dF1=pd.DataFrame(data)
dF1.head()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver import Keys
from selenium.webdriver.common.by import By
import time

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

driver = webdriver.Chrome()
driver.get("https://www.1mg.com/categories/homeopathy-57")
driver.maximize_window()
# time.sleep(3)

cancel_b = driver.find_element(By.XPATH, '//div[@class="UpdateCityModal__cancel-btn___2jWwS UpdateCityModal__btn___oMW5n"]')
if cancel_b.is_displayed():
    cancel_b.click()
    # time.sleep(2)


html=driver.find_element(By.TAG_NAME,'html')
for i in range(5):
    html.send_keys(Keys.PAGE_DOWN)
    
product_links = []
page=1
while page <= 2:
    product_elements = driver.find_elements(By.XPATH, '//div[@class="style__product-box___liepi"]/a')
    for product_element in product_elements:
        link = product_element.get_attribute('href')
        product_links.append(link)


    page+=1
    next_b_b = driver.find_element(By.XPATH, '//li[@class="next"]/a')
    if next_b_b.is_displayed():
        next_b_b.click()
    time.sleep(2)


print(len(product_links))
Ingredients = []

for product_link in product_links[:100]:  # Adjust the range as needed
    var = {
        'Origin': 'https://www.1mg.com',
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
    }

    response = requests.get(product_link, headers=var)
    
    if response.status_code == 404:
        print("break")
        break
    else:
        soup = BeautifulSoup(response.text, "lxml")
        ingredient = soup.find("div", {"class": "ProductDescription__description-content___A_qCZ"})

        if ingredient:
            Ingredients.append(ingredient.text.strip())
        else:
            Ingredients.append("NA")

print(Ingredients)
```
